**Para realizar la visualización de los datos tomaré como referencia los últimos 5 años con saltos de 3 meses.
Para el análisis usaré boxplot para visualizar outliers y gráfico de barras**

```python
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

APPLE

```python
ap = apple.history(period='5y',interval='3mo')
ap.reset_index(inplace=True)
ap
```

```python
sns.barplot(x= 'Close', y='Date', data=ap)

## El primer hallazgo es que, desde junio/2020, momento en que inició la pandemia de COVID19, 
# el precio de cierre de las acciones de Apple aumentó en casi 50%: pasó de 80USD a 120USD. 
## El aumento en la cotización de cierre ha tenido un crecimiendo sostenido desde el año 2020.
## Se puede ver que, en aproximadamente 3 años, la cotización aumentó en casi 100%. 
```

```python
sns.boxplot(x='Close', data=ap, color='grey')

## En este gráfico se puede observar una Mediana de 120USD, la cual está más cerca del valor 
# máximo de la cotización. Esto implicaría que, habría la misma cantidad de datos por debajo
# y por encima, indicando que desde que la acción vale 120USD habría menos variación de precio. 
## La mayor concentración de datos estaría en un rango de 55USD a 145USD. Por debajo y por encima,
# se encontraría los valores atípicos. 
```

MICROSOFT

```python
ms = microsoft.history(period='5y',interval='3mo')
ms.reset_index(inplace=True)
ms
```

```python
sns.barplot(x= 'Close', y='Date', data=ms)

## Se puede observar que en el mes de septiembre de 2021, la cotización alcanzó un valor atípico
# de aproximadamente 350USD, y luego tuvo una tendencia bajista. 
## La compañía inició este año, 2023, con un aumento en la cotización, y actualmente rondaría 280USD.
```

```python
sns.boxplot(x='Close', data=ms, color='blue')

## En este gráfico se puede observar una mediana de 220USD.
## La mayoría de las cotizaciones se encuentran en el rango de 130USD a 260USD, el resto serían
# valores atípicos. 
```

```python
sns.barplot(x= 'Dividends', y='Date', data=ms)

## Agregué la variable 'Dividendos' ya que resultó llamativo ver que esta compañía 
# paga dividendos cada trimestre, desde los últimos 5 años. 
# El valor de los mismos ha aumentado de manera anual.
```

WALMART

```python
wm = walmart.history(period='5y',interval='3mo')
wm.reset_index(inplace=True)
wm
```

```python
sns.barplot(x= 'Close', y='Date', data=wm)

## El valor máximo de la cotización fue alcanzado en septiembre 2022, y anteriormente en septiembre 2020,
# momento de pandemia por COVID19.
## Se muestra que en estos últimos 5 años las acciones han aumentado apenas 75%. 
## Desde marzo 2020 la cotización subió a 120USD y mantuvo su tendencia alcista. 
```

```python
sns.boxplot(x='Close', data=wm, color='purple')

## En este caso se muestra que los valores típicos están dentro del rango de 100USD a 140USD aproximadamente.
## El precio de 150USD es atípico. 
```

```python
sns.barplot(x= 'Dividends', y='Date', data=wm)

## Agregué la variable 'Dividendos' ya que resultó llamativo ver que esta compañía 
# no paga dividendos en el segundo semestre del año (junio a noviembre), pero sí 
# lo hace en los meses de marzo, junio y diciembre.  
## En los meses de marzo de cada año se recibe el mayor porcentaje, ya que sería un poco más de 1%,
# mientras que el resto de los meses se recibe aprox 0,6%.
```

CHEVRON

```python
ch = chevron.history(period='5y',interval='3mo')
ch.reset_index(inplace=True)
ch
```

```python
sns.barplot(x= 'Close', y='Date', data=ch)

## A simple vista se puede observar que esta compañía ha tenido poca volatilidad en el precio de sus acciones.
## En diciembre 2021 se puede notar un aumento de casi 40%, y desde allí su tendencia alcista. 
## En septiembre 2022 tuvo su pico de más de 175USD, pero en la actualidad su valor está en 155USD aproximadamente. 
```

```python
sns.boxplot(x='Close', data=ch, color='red')

## En este gráfico se pueden observar que los valores típicos están en el rango de 90USD a 
# 140USD aproximadamente. 
## Resulta llamativa la posición de la mediana, ubicada cerca del límite inferior de dicho rango, 
# en el valor de 100USD. Esto significaría que el precio ha sido volátil positivamente, ya que ha aumentado. 
```

```python
sns.barplot(x= 'Dividends', y='Date', data=ch)

## Agregué la variable 'Dividendos' para encontrar que esta compañía abona dividendos
# cada trimestre, desde los últimos 5 años. 
## El valor de los mismos aumenta en los meses de diciembre de cada año.
## Históricamente, en los últimos 5 años, el valor ha sido mayor al 1%. 
```
